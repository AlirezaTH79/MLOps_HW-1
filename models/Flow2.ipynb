# Parsbert + LDA + Logestic & random forest & xgboost + downsampling

```python
import torch
import pandas as pd
import numpy as np

# Load the PyTorch file
data = torch.load('listMain.pt')

# Convert it into a numpy array
numpy_data = np.array(data).reshape((32728, 768))

# Convert the numpy array into a pandas dataframe
df_bert = pd.DataFrame(numpy_data)
```

```python
df_bert
```

```python
df = pd.read_csv('Dataset_v2.csv')
```

```python
df
```

```python
name_list = df['cat1'].tolist()

df_bert.insert(0, "label", name_list)
```

```python
df_bert
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

encoder = LabelEncoder()
y_features = encoder.fit_transform(df_bert["label"])

X_train, X_test, y_train, y_test = train_test_split(df_bert.drop(columns=["label"]), y_features, test_size=0.2)


LDA_transform = LDA(n_components=9)
LDA_transform.fit(X_train, y_train)
features_new_train = LDA_transform.transform(X_train)
features_new_test = LDA_transform.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


classifier = LogisticRegression(max_iter=10000)

# Train the classifier on the training data
classifier.fit(features_new_train, y_train)

# Make predictions on the test data
y_pred = classifier.predict(features_new_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split


# Create the Random Forest classifier object with default hyperparameters
rf_classifier = RandomForestClassifier()

# Fit the classifier to the training data
rf_classifier.fit(features_new_train, y_train)

# Predict labels for the test data
y_pred = rf_classifier.predict(features_new_test)

# Evaluate the classifier's performance on the test data
accuracy = rf_classifier.score(features_new_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
import xgboost as xgb


# Train an XGBoost classifier for multi-class classification
xgb_classifier = xgb.XGBClassifier(objective="multi:softmax", num_class=10, n_estimators=100, max_depth=10)
xgb_classifier.fit(features_new_train, y_train)

# Use the trained model to make predictions on the test set
y_pred = xgb_classifier.predict(features_new_test)

# Calculate the accuracy of the model on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
df_new = df_bert[df_bert['label'] != 'اجتماعی'].reset_index(drop=True)
```

```python
categories = {}

for i in range(df_new.shape[0]):
    if df_new.loc[i, 'label'] not in categories:
        categories[df_new.loc[i, 'label']] = 1
    else:
         categories[df_new.loc[i, 'label']] += 1
```

```python
categories
```

```python
import pandas as pd


# Group the dataframe by the "label" column
groups = df_new.groupby('label')

# Calculate the minimum count value among all labels
min_count = groups.count().min()[0]

# Sample the same number of rows for each label using the minimum count value
subsets = []
for _, group in groups:
    subset = group.sample(n=min_count)
    subsets.append(subset)

# Concatenate the selected subsets back into a single dataframe
result = pd.concat(subsets).sample(frac=1).reset_index(drop=True)
```

```python
result
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

encoder = LabelEncoder()
y_features = encoder.fit_transform(result["label"])

X_train, X_test, y_train, y_test = train_test_split(result.drop(columns=["label"]), y_features, test_size=0.2)


LDA_transform = LDA(n_components=8)
LDA_transform.fit(X_train, y_train)
features_new_train = LDA_transform.transform(X_train)
features_new_test = LDA_transform.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


classifier = LogisticRegression(max_iter=10000)

# Train the classifier on the training data
classifier.fit(features_new_train, y_train)

# Make predictions on the test data
y_pred = classifier.predict(features_new_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split


# Create the Random Forest classifier object with default hyperparameters
rf_classifier = RandomForestClassifier()

# Fit the classifier to the training data
rf_classifier.fit(features_new_train, y_train)

# Predict labels for the test data
y_pred = rf_classifier.predict(features_new_test)

# Evaluate the classifier's performance on the test data
accuracy = rf_classifier.score(features_new_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
import xgboost as xgb


# Train an XGBoost classifier for multi-class classification
xgb_classifier = xgb.XGBClassifier(objective="multi:softmax", num_class=9, n_estimators=100, max_depth=10)
xgb_classifier.fit(features_new_train, y_train)

# Use the trained model to make predictions on the test set
y_pred = xgb_classifier.predict(features_new_test)

# Calculate the accuracy of the model on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
