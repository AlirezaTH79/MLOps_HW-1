# FastText & TSNE & Logestic & random forest & xgboost & weigthed loss logestic regression

```python
import pandas as pd


df = pd.read_csv('Dataset_v2.csv')
TSNE_fast_text_df = pd.read_csv('TSNE_fast_text.csv')
```

```python
TSNE_fast_text_df = TSNE_fast_text_df.drop(TSNE_fast_text_df.columns[0], axis=1)
TSNE_fast_text_df
```

```python
name_list = df['cat1'].tolist()

TSNE_fast_text_df.insert(0, "label", name_list)
```

```python
TSNE_fast_text_df
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

encoder = LabelEncoder()
y_features = encoder.fit_transform(TSNE_fast_text_df["label"])
X_train, X_test, y_train, y_test = train_test_split(TSNE_fast_text_df.drop(columns=["label"]), y_features, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


classifier = LogisticRegression(max_iter=10000)

# Train the classifier on the training data
classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split


# Create the Random Forest classifier object with default hyperparameters
rf_classifier = RandomForestClassifier()

# Fit the classifier to the training data
rf_classifier.fit(X_train, y_train)

# Predict labels for the test data
y_pred = rf_classifier.predict(X_test)

# Evaluate the classifier's performance on the test data
accuracy = rf_classifier.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
import xgboost as xgb


# Train an XGBoost classifier for multi-class classification
xgb_classifier = xgb.XGBClassifier(objective="multi:softmax", num_class=10, n_estimators=100, max_depth=10)
xgb_classifier.fit(X_train, y_train)

# Use the trained model to make predictions on the test set
y_pred = xgb_classifier.predict(X_test)

# Calculate the accuracy of the model on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# set weights for the two classes
weights = {}
for i in range(10):
    count = 0
    for j in y_features:
        if j == i:
            count += 1
    weights[i] = count / len(y_features)

# initialize the logistic regression model with weight balancing
clf = LogisticRegression(class_weight=weights)

# fit the model to the data
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
