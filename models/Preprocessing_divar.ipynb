```python
columns = []
for i in data:
    for j in i:
        if j not in columns:
            columns += [j]
```

```python
len(columns)
```

```python
data_dict = {}
for i in columns:
    data_dict[i] = []
```

```python
for i in data:
    keysList = list(i.keys())
    for j in columns:
        if j in keysList:
            data_dict[j] += [i[j]]
        else:
            data_dict[j] += ['NaN']
```

```python
df_v1 = pd.DataFrame.from_dict(data_dict)
```
