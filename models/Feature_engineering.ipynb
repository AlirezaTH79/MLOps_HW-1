```python
descriptions = []

for i in range(df.shape[0]):
    if str(df.loc[i, 'description2']) != 'nan':
        descriptions += [df.loc[i, 'description'] + df.loc[i, 'description2']]
    else:
        descriptions += [df.loc[i, 'description']]
```

```python
import re

temp = ['_', '__', '___', '____', '______', '__________', '___________',
       '____________', '_____________', '______________',
       '_______________', '________________', '_________________',
       '__________________', '___________________',
       '____________________', '_____________________',
       '______________________', '_______________________',
       '________________________', '_________________________',
       '__________________________', '___________________________',
       '____________________________', '_____________________________',
       '______________________________',
       '_______________________________',
       '________________________________',
       '__________________________________',
       '___________________________________',
       '____________________________________',
       '_____________________________________',
       '______________________________________',
       '_______________________________________',
       '________________________________________',
       '____________________________________________',
       '_____________________________________________',
       '______________________________________________',
       '_______________________________________________',
       '✅', '⬆️', '♥', '⬇️', '⚪️']

def remove_emojis(text):
    # Emoji ranges taken from https://unicode.org/emoji/charts/full-emoji-list.html
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

REPLACEMENTS = []

for i in temp:
    REPLACEMENTS += [(i, '')]

for old, new in REPLACEMENTS:
    for i in range(len(descriptions)):
        descriptions[i] = remove_emojis(descriptions[i].replace(old, new))
```

```python
from __future__ import unicode_literals
from hazm import *
import random 
import numpy as np
import nltk
import pandas as pd
import codecs
import tqdm

hazm_normalizer = Normalizer()

descriptions_normalized_hazm = [hazm_normalizer.normalize(x) for x in tqdm.tqdm(descriptions)]
```

```python
descriptions_sentences = [sent_tokenize(x) for x in tqdm.tqdm(descriptions_normalized_hazm)]
```

```python
descriptions_tokens = [[word_tokenize(sent) for sent in sents] for sents in tqdm.tqdm(descriptions_sentences)]
```

```python
stop_path = '/content/drive/MyDrive/MLops/stopwords.txt'
stopwords = [hazm_normalizer.normalize(x.strip()) for x in codecs.open(stop_path,'r','utf-8').readlines()]
```

```python
descriptions_final = []

c = 0
for i in descriptions_tokens:
    c += 1
    if c % 1000 == 0:
        print(c)
    text = ''
    for j in i:
        for k in j:
            if k not in stopwords:
                text += k 
                text += ' '
    
    descriptions_final += [text]
```

```python
from transformers import AutoConfig, AutoTokenizer, AutoModel

config = AutoConfig.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
tokenizer = AutoTokenizer.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
model = AutoModel.from_pretrained("HooshvareLab/bert-base-parsbert-uncased" )
```

```python
tokens = [tokenizer.encode(i, add_special_tokens=True) for i in descriptions]
ids = pad_sequence(tokens, maxlen=400, dtype='long', value=0, truncating='post', padding='post')
masks = []
for i in ids:
    masks += [[int(j > 0) for j in i]

ids = torch.tensor(ids).to('cuda:0')
masks = torch.tensor(masks).to('cuda:0')
```

```python
model.to('cuda:0')

bert_features = []

with torch.no_grad():
    last_hidden_states = model(ids, attention_mask=masks)
    bert_features += [last_hidden_states[0][:,0,:]] 
```

# Fast text

```python
from gensim.models.fasttext import FastText

fast_text = FastText(descriptions_final,
                     vector_size=250,
                     window=5,
                     min_count=5,
                     sample=5e-2,
                     workers=5,
                     sg=1,
                     epochs=25)
```

```python
words_list = list(tr_idf_model.get_feature_names_out())

fast_text_emb = []
for idx in range(len(descriptions_final)):
    comment_tfidf = tf_idf_vector[idx]
    comment_emb = []
    for token in descriptions_final[idx]:
        token_fast_emb = fast_text.wv[token]
        token_final_emb = token_fast_emb * comment_tfidf[0 , words_list.index(token)]
        comment_emb.append(token_final_emb)
            
    fast_text_emb.append(np.mean(comment_emb,axis=0))
```

```python
fast_text_df = pd.dataframe(fast_text_emb)
```

```python
fast_text_df.save('fast_text.csv')
```
